```python
import pandas as pd

df = pd.read_csv('accident.csv')

print(df.head(10))
```

```python
import matplotlib.pyplot as plt

severity_counts = df['Severity'].value_counts()
severity_counts.plot(kind='bar', color='skyblue')

plt.title('Accident Severity Distribution')
plt.xlabel('Severity')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
weather_counts = df['WeatherConditions'].value_counts()
weather_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140)

plt.title('Weather Conditions During Accidents')
plt.ylabel('')  
plt.show()
```

```python
day_counts = df['Day'].value_counts().sort_index()

day_counts.plot(kind='line', marker='o', color='lightgreen')

plt.title('Accidents by Day of the Week')
plt.xlabel('Day')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
plt.scatter(df['SpeedLimit'], df['Severity'], alpha=0.5, color='red')

plt.title('Speed Limit vs. Accident Severity')
plt.xlabel('Speed Limit')
plt.ylabel('Severity')
plt.show()
```

```python
df['SpeedLimit'].plot(kind='hist', bins=15, color='purple', edgecolor='black')

plt.title('Distribution of Speed Limits')
plt.xlabel('Speed Limit')
plt.ylabel('Frequency')
plt.show()
```
